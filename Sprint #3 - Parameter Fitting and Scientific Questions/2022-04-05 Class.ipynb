## SIR Model

```python
from pyndamics3 import Simulation
```

```python
sim=Simulation()
sim.add("S' = -β*S*I/N",100,plot=1)
sim.add("I' = +β*S*I/N - γ*I",1,plot=1)
sim.add("R' = +γ*I",0,plot=1)
sim.params(β=2,γ=1,N=101)
sim.run(10)
```

```python
sim=Simulation()
sim.add("S' = -β*S*I/N",100,plot=1)
sim.add("I' = +β*S*I/N - γ*I",1,plot=1)
sim.add("R' = +γ*I",0,plot=1)
sim.add("N = S+I+R",plot=1)
sim.params(β=2,γ=1)
sim.run(10)
```

```python
x=sim.t
y=sim.I

plot(x,y)
```

```python
x=sim.t
y=max(sim.I)-sim.I

plot(x,y)
```
