```python
def randu(*args,seed=None):
    from numpy import prod,array
    if len(args)==0:
        global Ij
        
        if not seed is None:
            Ij=seed
            return
        
        try:
            _ = Ij
        except NameError:
            Ij=101

        m=2**31
        c=0
        a=65539

        Ij=(a*Ij+c) % m


        return Ij/m
    
    else:
        N=prod(args)
        a=array([randu() for _ in range(N)]).reshape(*args)
        return a
```
