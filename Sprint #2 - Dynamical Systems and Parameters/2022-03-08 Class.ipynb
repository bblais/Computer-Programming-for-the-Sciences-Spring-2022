```python
from pyndamics3 import Simulation
```

```python
sim=Simulation()
sim.add("N'=r*N*(1-N/K)-H",1,plot=True)
sim.params(r=1,K=50,H=.979)
sim.run(600)
```

```python
sim=Simulation()
sim.add("x'=v",1)
sim.add("v'=a",0)
sim.add("a=k*vm*exp(-k*t)")
sim.params(k=3,vm=2)
sim.run(600)
```
