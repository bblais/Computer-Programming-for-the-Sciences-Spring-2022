```python
from pyndamics3 import Simulation
from sci378 import *
```

$$\frac{dx}{dt}=ax\left(1-\frac{x}{K}\right)$$

```python
sim=Simulation()
sim.add("x' = a*x*(1-x/K)",1,plot=True)
sim.params(a=1,K=50)
sim.run(10)
```

$$\frac{dx}{dt}= v$$

$$\frac{dv}{dt}= -k\cdot x/m$$

```python
sim=Simulation()
sim.add("x'=v",1)
sim.add("v'=-k*x/m",0)
sim.params(k=2,m=1)
sim.run(10)
```

```python
subplot(2,1,1)
plot(sim.t,sim.x)
ylabel('x')
gca().set_xticklabels([])
subplot(2,1,2)
plot(sim.t,sim.v)
ylabel('v')
xlabel('t')
```

```python
sim=Simulation()
sim.add("x'=α*x-β*x*y",10,plot=1)
sim.add("y'=δ*x*y-γ*y",10,plot=1)
sim.params(α=1.1,β=0.4,δ=.1,γ=.4)
sim.run(100)
```

```python
plot(sim.t,sim.x,label='Baboons')
plot(sim.t,sim.y,label='Cheetahs')
legend(loc='upper left')
ylim([0,37])
```
